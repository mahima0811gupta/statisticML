Dot product

```python
a=np.asarray([2,3])
b=np.asarray([4,4])
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

Cross Product of two vector

```python
a=np.asarray([2,3])
b=np.asarray([4,4])
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

```python
a=np.asarray([5,10,20])
b=np.asarray([18,32,50])
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

Projection of a vector on v vector

```python
import numpy as np

a = np.asarray([2, 5])
v = np.asarray([8, -6])

# Magnitude of v vector
magnitude_of_v = np.sqrt(np.sum(v**2))

# Projection of a on v
proj_of_a_on_v = (np.dot(a, v) / magnitude_of_v**2) * v

print("Projection of a on v:", proj_of_a_on_v)
```
